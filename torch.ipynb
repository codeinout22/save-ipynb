```python
import torch
```

# 파이토치의 여러 함수들

```python
A = torch.randn(3,3) # Normal의 n
B = torch.rand(3,3) # Uniform

print(A)
print(A[:, 0]<0)
print(A[A[:, 0]<0, :], end="\n\n")

print(B)
```

```python
A = torch.randn(3,3)
print(A)
print(torch.abs(A)) #절댓값
print(torch.sqrt(torch.abs(A)))
```

```python
# sqrt 제곱근 구하는 함수 (squate root)
torch.sqrt(torch.tensor(4))
```

```python
print(torch.exp(A))
print(torch.log(torch.abs(A)))
print(torch.log2(torch.tensor(2)))
print(torch.log10(torch.tensor(10)))
```
